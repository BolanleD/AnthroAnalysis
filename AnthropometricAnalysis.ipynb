```r
maleData <- read.csv("Male.csv", header = TRUE)
femaleData <- read.csv("Female.csv", header = TRUE)
```

```r
# combine by row
fullData <- rbind(maleData, femaleData)
```

```r
library(ggplot2)
fullData <- rbind(maleData, femaleData)
```

```r
fullData <- transform(fullData, BMIndex = (Weightlbs/(Heightin*Heightin))*703)
```

```r
Compare <- ggplot(fullData, aes(x=Age, y=BMIndex, color=Gender)) + geom_point()
print(Compare)
```

```r
CompareWaist <- ggplot(fullData, aes(x=waistcircumference, y=BMIndex, color=Gender)) + geom_point()
print(CompareWaist)
```
